````python
# 🔬 Micro Practice: Implement MQA and GQA
# Goal: Build efficient attention variants
# Expected outcome: Demonstrate memory savings

class MultiHeadAttention(nn.Module):
    """Standard Multi-Head Attention"""
    
    def __init__(self, d_model, num_heads):
        super().__init__()
        self.d_model = d_model
        self.num_heads = num_heads
        self.head_dim = d_model // num_heads
        
        self.q_proj = nn.Linear(d_model, d_model)
        self.k_proj = nn.Linear(d_model, d_model)
        self.v_proj = nn.Linear(d_model, d_model)
        self.out_proj = nn.Linear(d_model, d_model)
    
    def forward(self, x):
        batch_size, seq_len, d_model = x.shape
        
        # Project and reshape
        q = self.q_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)
        k = self.k_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)
        v = self.v_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)
        
        # Transpose for attention
        q = q.transpose(1, 2)  # [batch, num_heads, seq_len, head_dim]
        k = k.transpose(1, 2)
        v = v.transpose(1, 2)
        
        # Attention
        scores = torch.matmul(q, k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn = torch.softmax(scores, dim=-1)
        out = torch.matmul(attn, v)
        
        # Reshape and project
        out = out.transpose(1, 2).contiguous().view(batch_size, seq_len, d_model)
        out = self.out_proj(out)
        
        return out, (k, v)  # Return KV for cache size comparison

class MultiQueryAttention(nn.Module):
    """Multi-Query Attention (shared K, V across heads)"""
    
    def __init__(self, d_model, num_heads):
        super().__init__()
        self.d_model = d_model
        self.num_heads = num_heads
        self.head_dim = d_model // num_heads
        
        self.q_proj = nn.Linear(d_model, d_model)
        self.k_proj = nn.Linear(d_model, self.head_dim)  # Single K
        self.v_proj = nn.Linear(d_model, self.head_dim)  # Single V
        self.out_proj = nn.Linear(d_model, d_model)
    
    def forward(self, x):
        batch_size, seq_len, d_model = x.shape
        
        # Project Q (multiple heads)
        q = self.q_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)
        q = q.transpose(1, 2)  # [batch, num_heads, seq_len, head_dim]
        
        # Project K, V (single head, shared)
        k = self.k_proj(x).unsqueeze(1)  # [batch, 1, seq_len, head_dim]
        v = self.v_proj(x).unsqueeze(1)  # [batch, 1, seq_len, head_dim]
        
        # Expand K, V to match Q
        k = k.expand(-1, self.num_heads, -1, -1)
        v = v.expand(-1, self.num_heads, -1, -1)
        
        # Attention
        scores = torch.matmul(q, k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn = torch.softmax(scores, dim=-1)
        out = torch.matmul(attn, v)
        
        # Reshape and project
        out = out.transpose(1, 2).contiguous().view(batch_size, seq_len, d_model)
        out = self.out_proj(out)
        
        return out, (k[:, 0:1, :, :], v[:, 0:1, :, :])  # Return single KV

class GroupedQueryAttention(nn.Module):
    """Grouped-Query Attention (grouped K, V)"""
    
    def __init__(self, d_model, num_heads, num_kv_groups):
        super().__init__()
        self.d_model = d_model
        self.num_heads = num_heads
        self.num_kv_groups = num_kv_groups
        self.head_dim = d_model // num_heads
        
        assert num_heads % num_kv_groups == 0, "num_heads must be divisible by num_kv_groups"
        
        self.q_proj = nn.Linear(d_model, d_model)
        self.k_proj = nn.Linear(d_model, num_kv_groups * self.head_dim)
        self.v_proj = nn.Linear(d_model, num_kv_groups * self.head_dim)
        self.out_proj = nn.Linear(d_model, d_model)
    
    def forward(self, x):
        batch_size, seq_len, d_model = x.shape
        
        # Project Q
        q = self.q_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim)
        q = q.transpose(1, 2)  # [batch, num_heads, seq_len, head_dim]
        
        # Project K, V (grouped)
        k = self.k_proj(x).view(batch_size, seq_len, self.num_kv_groups, self.head_dim)
        v = self.v_proj(x).view(batch_size, seq_len, self.num_kv_groups, self.head_dim)
        k = k.transpose(1, 2)  # [batch, num_kv_groups, seq_len, head_dim]
        v = v.transpose(1, 2)
        
        # Repeat K, V for each group
        heads_per_group = self.num_heads // self.num_kv_groups
        k = k.repeat_interleave(heads_per_group, dim=1)
        v = v.repeat_interleave(heads_per_group, dim=1)
        
        # Attention
        scores = torch.matmul(q, k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn = torch.softmax(scores, dim=-1)
        out = torch.matmul(attn, v)
        
        # Reshape and project
        out = out.transpose(1, 2).contiguous().view(batch_size, seq_len, d_model)
        out = self.out_proj(out)
        
        return out, (k[:, :self.num_kv_groups, :, :], v[:, :self.num_kv_groups, :, :])

# Compare attention variants
print("=" * 60)
print("高效注意力变体对比")
print("=" * 60)

d_model = 512
num_heads = 8
num_kv_groups = 2
seq_len = 1024

# Create models
mha = MultiHeadAttention(d_model, num_heads)
mqa = MultiQueryAttention(d_model, num_heads)
gqa = GroupedQueryAttention(d_model, num_heads, num_kv_groups)

# Test input
batch_size = 1
x = torch.randn(batch_size, seq_len, d_model)

# Forward pass
out_mha, (k_mha, v_mha) = mha(x)
out_mqa, (k_mqa, v_mqa) = mqa(x)
out_gqa, (k_gqa, v_gqa) = gqa(x)

# Calculate KV cache sizes
kv_cache_mha = k_mha.numel() + v_mha.numel()
kv_cache_mqa = k_mqa.numel() + v_mqa.numel()
kv_cache_gqa = k_gqa.numel() + v_gqa.numel()

print(f"\n配置：")
print(f"  模型维度: {d_model}")
print(f"  注意力头数: {num_heads}")
print(f"  KV 组数 (GQA): {num_kv_groups}")
print(f"  序列长度: {seq_len}")

print(f"\nKV 缓存大小对比：")
print(f"  MHA: {kv_cache_mha:,} 元素 (1.00x)")
print(f"  MQA: {kv_cache_mqa:,} 元素 ({kv_cache_mqa/kv_cache_mha:.2f}x)")
print(f"  GQA: {kv_cache_gqa:,} 元素 ({kv_cache_gqa/kv_cache_mha:.2f}x)")

print(f"\n内存节省：")
print(f"  MQA vs MHA: {(1 - kv_cache_mqa/kv_cache_mha)*100:.1f}% 减少")
print(f"  GQA vs MHA: {(1 - kv_cache_gqa/kv_cache_mha)*100:.1f}% 减少")

# Visualize KV cache sizes
methods = ['MHA', 'MQA', 'GQA']
cache_sizes = [kv_cache_mha, kv_cache_mqa, kv_cache_gqa]
colors = ['#667eea', '#f093fb', '#4facfe']

plt.figure(figsize=(10, 6))
bars = plt.bar(methods, cache_sizes, color=colors, alpha=0.7)
plt.ylabel('KV Cache Size (elements)', fontsize=12)
plt.title('KV Cache Memory Comparison', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

# Add value labels
for i, (method, size) in enumerate(zip(methods, cache_sizes)):
    plt.text(i, size, f'{size/1e6:.2f}M\n({size/kv_cache_mha:.2f}x)', 
             ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

print("\n✓ 高效注意力变体实现完成！")
print("\n关键洞察：")
print("  - MQA: 最大内存节省，质量略有损失")
print("  - GQA: 平衡质量和效率")
print("  - 推理时内存是瓶颈，这些优化很重要")
````

## 7. 高效注意力架构 (Efficient Architectures)

### 7.1 注意力效率优化

**标准多头注意力 (MHA) 的问题**：
- 每个头都有独立的 K, V 投影
- KV 缓存占用大量内存
- 推理时内存成为瓶颈

### 7.2 Multi-Query Attention (MQA)

**核心思想**：所有头共享同一组 K, V。

**标准 MHA**：
```
Q: [batch, seq_len, num_heads, head_dim]
K: [batch, seq_len, num_heads, head_dim]
V: [batch, seq_len, num_heads, head_dim]
```

**MQA**：
```
Q: [batch, seq_len, num_heads, head_dim]
K: [batch, seq_len, 1, head_dim]  ← 共享
V: [batch, seq_len, 1, head_dim]  ← 共享
```

**优势**：
- KV 缓存减少 $n_{\text{heads}}$ 倍
- 推理速度提升 1.5-2x
- 质量损失 < 1%

**应用**：
- PaLM (Google)
- Falcon (TII)

### 7.3 Grouped-Query Attention (GQA)

**核心思想**：MHA 和 MQA 的折中，将头分组共享 K, V。

**GQA**：
```
Q: [batch, seq_len, num_heads, head_dim]
K: [batch, seq_len, num_groups, head_dim]  ← 分组共享
V: [batch, seq_len, num_groups, head_dim]  ← 分组共享
```

其中 `num_groups < num_heads`。

**例子**：
- 8 个查询头，2 个 KV 组
- 每 4 个查询头共享一组 KV
- KV 缓存减少 4 倍

**优势**：
- 平衡质量和效率
- 比 MQA 质量更好
- 比 MHA 更高效

**应用**：
- LLaMA-2 (Meta)
- Mistral (Mistral AI)

### 7.4 Flash Attention

**核心思想**：IO 感知的注意力算法。

**标准注意力的问题**：
- 需要实例化完整的注意力矩阵 $O(n^2)$
- 频繁的 HBM ↔ SRAM 数据传输
- IO 成为瓶颈

**Flash Attention 优化**：
- 分块计算（tiling）
- 在 SRAM 中完成计算
- 减少 HBM 访问
- 速度提升 2-4x，内存减少 10-20x

**Flash Attention 2**：
- 进一步优化并行性
- 更好的工作分配
- 速度再提升 2x

### 7.5 Sliding Window Attention

**核心思想**：只关注固定窗口内的 token。

**优势**：
- 复杂度：$O(n \cdot w)$ 而非 $O(n^2)$
- 适合长序列
- 可以堆叠多层扩大感受野

**应用**：
- Mistral 7B：窗口大小 4096
- Longformer：可变窗口

### 7.6 效率对比

| 方法 | KV 缓存 | 速度 | 质量 | 实现难度 |
|------|---------|------|------|---------|
| **MHA** | 1x | 1x | 100% | 简单 |
| **MQA** | 1/n | 1.5-2x | 98-99% | 简单 |
| **GQA** | 1/g | 1.2-1.5x | 99-100% | 简单 |
| **Flash Attn** | 1x | 2-4x | 100% | 复杂 |
| **Sliding Window** | 1x | n/w | 95-98% | 中等 |

其中 $n$ 是头数，$g$ 是组数，$w$ 是窗口大小。

````python
# 🔬 Micro Practice: Implement Simple Vision-Language Model
# Goal: Build model combining vision encoder with language model
# Expected outcome: Multimodal fusion architecture

class SimpleVisionEncoder(nn.Module):
    """Simple CNN-based vision encoder"""
    
    def __init__(self, image_size=224, patch_size=16, d_model=512):
        super().__init__()
        self.image_size = image_size
        self.patch_size = patch_size
        self.num_patches = (image_size // patch_size) ** 2
        
        # Patch embedding (simplified CNN)
        self.patch_embed = nn.Conv2d(
            3, d_model, 
            kernel_size=patch_size, 
            stride=patch_size
        )
        
        # Position embeddings
        self.pos_embed = nn.Parameter(torch.randn(1, self.num_patches, d_model) * 0.02)
    
    def forward(self, x):
        """
        Args:
            x: Images [batch_size, 3, image_size, image_size]
        
        Returns:
            features: Patch features [batch_size, num_patches, d_model]
        """
        # Extract patches
        x = self.patch_embed(x)  # [batch_size, d_model, H, W]
        x = x.flatten(2).transpose(1, 2)  # [batch_size, num_patches, d_model]
        
        # Add position embeddings
        x = x + self.pos_embed
        
        return x

class VisionLanguageModel(nn.Module):
    """
    Simple Vision-Language Model with Cross-Attention
    
    Architecture:
        Image → Vision Encoder → Image Features
                                      ↓ Cross-Attention
        Text → Text Encoder → Text Features → Fused Features
    
    Args:
        d_model: Model dimension
        image_size: Input image size
        patch_size: Patch size for vision encoder
        vocab_size: Vocabulary size for text
    """
    
    def __init__(self, d_model=512, image_size=224, patch_size=16, vocab_size=10000):
        super().__init__()
        self.d_model = d_model
        
        # Vision encoder
        self.vision_encoder = SimpleVisionEncoder(image_size, patch_size, d_model)
        
        # Text embedding
        self.text_embed = nn.Embedding(vocab_size, d_model)
        self.text_pos_embed = nn.Parameter(torch.randn(1, 512, d_model) * 0.02)
        
        # Cross-attention layer
        self.cross_attn = nn.MultiheadAttention(d_model, num_heads=8, batch_first=True)
        
        # Output projection
        self.output_proj = nn.Linear(d_model, vocab_size)
    
    def forward(self, images, text_tokens):
        """
        Forward pass
        
        Args:
            images: Input images [batch_size, 3, H, W]
            text_tokens: Text token IDs [batch_size, seq_len]
        
        Returns:
            logits: Output logits [batch_size, seq_len, vocab_size]
        """
        batch_size, seq_len = text_tokens.shape
        
        # Encode image
        image_features = self.vision_encoder(images)  # [batch_size, num_patches, d_model]
        
        # Embed text
        text_features = self.text_embed(text_tokens)  # [batch_size, seq_len, d_model]
        text_features = text_features + self.text_pos_embed[:, :seq_len, :]
        
        # Cross-attention: text attends to image
        fused_features, attn_weights = self.cross_attn(
            query=text_features,
            key=image_features,
            value=image_features
        )
        
        # Output projection
        logits = self.output_proj(fused_features)
        
        return logits, attn_weights

# Create vision-language model
print("=" * 60)
print("视觉-语言模型实现")
print("=" * 60)

d_model = 512
image_size = 224
patch_size = 16
vocab_size = 10000

vl_model = VisionLanguageModel(d_model, image_size, patch_size, vocab_size)

# Test forward pass
batch_size = 2
images = torch.randn(batch_size, 3, image_size, image_size)
text_tokens = torch.randint(0, vocab_size, (batch_size, 20))

logits, attn_weights = vl_model(images, text_tokens)

print(f"\n输入：")
print(f"  图像形状: {images.shape}")
print(f"  文本形状: {text_tokens.shape}")

print(f"\n输出：")
print(f"  Logits 形状: {logits.shape}")
print(f"  注意力权重形状: {attn_weights.shape}")

# Count parameters
total_params = sum(p.numel() for p in vl_model.parameters())
vision_params = sum(p.numel() for p in vl_model.vision_encoder.parameters())
text_params = sum(p.numel() for p in vl_model.text_embed.parameters())

print(f"\n参数统计：")
print(f"  视觉编码器: {vision_params:,} 参数")
print(f"  文本嵌入: {text_params:,} 参数")
print(f"  总计: {total_params:,} 参数")

# Visualize cross-attention
attn_map = attn_weights[0].detach().numpy()  # [seq_len, num_patches]

plt.figure(figsize=(10, 6))
plt.imshow(attn_map, aspect='auto', cmap='viridis')
plt.colorbar(label='Attention Weight')
plt.xlabel('Image Patch', fontsize=12)
plt.ylabel('Text Token', fontsize=12)
plt.title('Cross-Attention: Text → Image', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n✓ 视觉-语言模型实现完成！")
print("\n关键特性：")
print("  - 视觉编码器提取图像特征")
print("  - 交叉注意力融合视觉和文本")
print("  - 文本可以关注图像的不同区域")
print("  - 适用于图像描述、VQA 等任务")
````

### 6.2 多模态架构类型

**1. 双编码器 (Dual Encoder)**

```
图像 → 视觉编码器 → 图像嵌入
                        ↓ 对比学习
文本 → 文本编码器 → 文本嵌入
```

**CLIP (Contrastive Language-Image Pre-training)**：
- 图像编码器：ViT 或 ResNet
- 文本编码器：Transformer
- 对比损失：匹配的图文对相似度高

$$L = -\log \frac{\exp(\text{sim}(i, t) / \tau)}{\sum_{j} \exp(\text{sim}(i, t_j) / \tau)}$$

**优点**：
- 简单高效
- 零样本迁移能力强
- 可扩展到大规模数据

**2. 交叉注意力 (Cross-Attention)**

```
图像特征 → [作为 K, V]
                ↓ 交叉注意力
文本特征 → [作为 Q] → 融合特征
```

**Flamingo**：
- 冻结的视觉编码器
- 冻结的语言模型
- 可训练的交叉注意力层
- 少样本学习能力强

**3. 统一架构 (Unified Architecture)**

```
图像 patch + 文本 token → 统一 Transformer → 输出
```

**GPT-4V / Gemini**：
- 将图像视为特殊 token

````python
# 🔬 Micro Practice: Implement RoPE Scaling
# Goal: Extend context length through positional interpolation
# Expected outcome: Better extrapolation to longer sequences

class RotaryPositionalEmbedding(nn.Module):
    """
    Rotary Position Embedding (RoPE)
    
    Args:
        dim: Dimension of embeddings
        max_seq_len: Maximum sequence length
        base: Base for frequency computation
        scaling_factor: Factor to scale positions (for interpolation)
    """
    
    def __init__(self, dim, max_seq_len=2048, base=10000, scaling_factor=1.0):
        super().__init__()
        self.dim = dim
        self.max_seq_len = max_seq_len
        self.base = base
        self.scaling_factor = scaling_factor
        
        # Compute frequencies
        inv_freq = 1.0 / (base ** (torch.arange(0, dim, 2).float() / dim))
        self.register_buffer('inv_freq', inv_freq)
        
        # Precompute cos and sin for max_seq_len
        self._update_cos_sin_cache(max_seq_len)
    
    def _update_cos_sin_cache(self, seq_len):
        """Precompute cos and sin values"""
        # Apply scaling to positions
        positions = torch.arange(seq_len).float() / self.scaling_factor
        
        # Compute angles
        freqs = torch.outer(positions, self.inv_freq)
        emb = torch.cat([freqs, freqs], dim=-1)
        
        self.register_buffer('cos_cached', emb.cos())
        self.register_buffer('sin_cached', emb.sin())
    
    def forward(self, x, seq_len=None):
        """
        Apply rotary embeddings
        
        Args:
            x: Input tensor [batch_size, seq_len, num_heads, head_dim]
            seq_len: Sequence length (if None, use x.shape[1])
        
        Returns:
            x_rotated: Tensor with rotary embeddings applied
        """
        if seq_len is None:
            seq_len = x.shape[1]
        
        # Get cached values
        cos = self.cos_cached[:seq_len, :]
        sin = self.sin_cached[:seq_len, :]
        
        # Apply rotation
        x_rotated = self._apply_rotary_emb(x, cos, sin)
        
        return x_rotated
    
    def _apply_rotary_emb(self, x, cos, sin):
        """Apply rotary transformation"""
        # Split into two halves
        x1, x2 = x[..., :self.dim//2], x[..., self.dim//2:]
        
        # Rotate
        x_rotated = torch.cat([
            x1 * cos[:, :self.dim//2] - x2 * sin[:, :self.dim//2],
            x1 * sin[:, self.dim//2:] + x2 * cos[:, self.dim//2:]
        ], dim=-1)
        
        return x_rotated

# Compare different scaling factors
print("=" * 60)
print("RoPE 位置编码插值")
print("=" * 60)

dim = 64
train_len = 2048
test_len = 4096  # 2x training length

# Create RoPE with different scaling factors
rope_no_scale = RotaryPositionalEmbedding(dim, max_seq_len=test_len, scaling_factor=1.0)
rope_linear_scale = RotaryPositionalEmbedding(dim, max_seq_len=test_len, scaling_factor=2.0)

print(f"\n训练长度: {train_len}")
print(f"测试长度: {test_len} (2x)")

# Test input
batch_size = 1
num_heads = 8
head_dim = dim
x = torch.randn(batch_size, test_len, num_heads, head_dim)

# Apply RoPE
x_no_scale = rope_no_scale(x)
x_linear_scale = rope_linear_scale(x)

print(f"\n输入形状: {x.shape}")
print(f"输出形状: {x_no_scale.shape}")

# Visualize frequency patterns
positions = torch.arange(test_len).float()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# No scaling
freqs_no_scale = torch.outer(positions, rope_no_scale.inv_freq)
ax1.imshow(freqs_no_scale[:512, :16].numpy(), aspect='auto', cmap='viridis')
ax1.set_xlabel('Frequency Dimension', fontsize=12)
ax1.set_ylabel('Position', fontsize=12)
ax1.set_title('RoPE without Scaling', fontsize=14, fontweight='bold')
ax1.axhline(y=train_len, color='red', linestyle='--', linewidth=2, label='Training length')
ax1.legend()

# Linear scaling
positions_scaled = positions / 2.0
freqs_linear_scale = torch.outer(positions_scaled, rope_linear_scale.inv_freq)
ax2.imshow(freqs_linear_scale[:512, :16].numpy(), aspect='auto', cmap='viridis')
ax2.set_xlabel('Frequency Dimension', fontsize=12)
ax2.set_ylabel('Position', fontsize=12)
ax2.set_title('RoPE with Linear Scaling (2x)', fontsize=14, fontweight='bold')
ax2.axhline(y=train_len, color='red', linestyle='--', linewidth=2, label='Training length')
ax2.legend()

plt.tight_layout()
plt.show()

print("\n缩放效果：")
print("  - 无缩放: 超出训练长度时频率模式未见过")
print("  - 线性缩放: 将位置压缩到训练范围内")
print("  - 结果: 更好的外推能力")

print("\n✓ RoPE 缩放实现完成！")
print("\n应用场景：")
print("  - 扩展预训练模型的上下文长度")
print("  - 无需重新训练")
print("  - LLaMA-2 使用此技术从 4K 扩展到 32K")
````

让我们实现 RoPE 缩放来扩展上下文长度。

````python
# 🔬 Micro Practice: Implement Sparse Attention (Sliding Window)
# Goal: Build sliding window + global attention pattern
# Expected outcome: Reduced complexity for long sequences

class SparseAttention(nn.Module):
    """
    Sparse Attention with Sliding Window + Global Tokens
    
    Pattern:
    - Each token attends to local window (w tokens on each side)
    - Special global tokens attend to all positions
    - Complexity: O(n * w) instead of O(n^2)
    
    Args:
        d_model: Model dimension
        window_size: Size of local attention window
        num_global_tokens: Number of tokens with global attention
    """
    
    def __init__(self, d_model, window_size=128, num_global_tokens=1):
        super().__init__()
        self.d_model = d_model
        self.window_size = window_size
        self.num_global_tokens = num_global_tokens
        
        self.qkv = nn.Linear(d_model, 3 * d_model)
        self.out = nn.Linear(d_model, d_model)
    
    def forward(self, x):
        """
        Forward pass with sparse attention
        
        Args:
            x: Input tensor [batch_size, seq_len, d_model]
        
        Returns:
            output: Attention output [batch_size, seq_len, d_model]
        """
        batch_size, seq_len, d_model = x.shape
        
        # Compute Q, K, V
        qkv = self.qkv(x)
        q, k, v = qkv.chunk(3, dim=-1)
        
        # Initialize output
        output = torch.zeros_like(x)
        
        # Process global tokens (first num_global_tokens)
        if self.num_global_tokens > 0:
            # Global tokens attend to everything
            q_global = q[:, :self.num_global_tokens, :]
            scores_global = torch.matmul(q_global, k.transpose(-2, -1)) / (d_model ** 0.5)
            attn_global = torch.softmax(scores_global, dim=-1)
            output[:, :self.num_global_tokens, :] = torch.matmul(attn_global, v)
        
        # Process local tokens with sliding window
        half_window = self.window_size // 2
        
        for i in range(self.num_global_tokens, seq_len):
            # Define window boundaries
            start = max(0, i - half_window)
            end = min(seq_len, i + half_window + 1)
            
            # Local attention within window
            q_i = q[:, i:i+1, :]  # [batch_size, 1, d_model]
            k_window = k[:, start:end, :]  # [batch_size, window_size, d_model]
            v_window = v[:, start:end, :]
            
            # Compute attention scores
            scores = torch.matmul(q_i, k_window.transpose(-2, -1)) / (d_model ** 0.5)
            attn = torch.softmax(scores, dim=-1)
            
            # Apply attention
            output[:, i:i+1, :] = torch.matmul(attn, v_window)
        
        # Output projection
        output = self.out(output)
        
        return output

# Create sparse attention layer
print("=" * 60)
print("稀疏注意力实现")
print("=" * 60)

d_model = 256
window_size = 128
num_global_tokens = 1

sparse_attn = SparseAttention(d_model, window_size, num_global_tokens)

# Test with long sequence
batch_size = 1
seq_len = 2048
x = torch.randn(batch_size, seq_len, d_model)

# Benchmark
start = time.time()
with torch.no_grad():
    output = sparse_attn(x)
sparse_time = (time.time() - start) * 1000

print(f"\n输入形状: {x.shape}")
print(f"输出形状: {output.shape}")
print(f"处理时间: {sparse_time:.2f} ms")

# Compare complexity
full_attn_ops = seq_len * seq_len * d_model
sparse_attn_ops = seq_len * window_size * d_model

print(f"\n复杂度对比：")
print(f"  完整注意力: {full_attn_ops:,} ops")
print(f"  稀疏注意力: {sparse_attn_ops:,} ops")
print(f"  减少: {full_attn_ops / sparse_attn_ops:.1f}x")

# Visualize attention pattern
def visualize_sparse_pattern(seq_len, window_size, num_global_tokens):
    """Visualize sparse attention pattern"""
    pattern = np.zeros((seq_len, seq_len))
    
    # Global tokens attend to everything
    pattern[:num_global_tokens, :] = 1
    
    # Local sliding window
    half_window = window_size // 2
    for i in range(num_global_tokens, seq_len):
        start = max(0, i - half_window)
        end = min(seq_len, i + half_window + 1)
        pattern[i, start:end] = 1
    
    return pattern

# Visualize for smaller sequence
vis_seq_len = 256
pattern = visualize_sparse_pattern(vis_seq_len, window_size, num_global_tokens)

plt.figure(figsize=(10, 8))
plt.imshow(pattern, cmap='Blues', aspect='auto')
plt.colorbar(label='Attention')
plt.xlabel('Key Position', fontsize=12)
plt.ylabel('Query Position', fontsize=12)
plt.title(f'Sparse Attention Pattern (Window={window_size})', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n✓ 稀疏注意力实现完成！")
print("\n关键特性：")
print("  - 滑动窗口局部注意力")
print("  - 全局 token 关注所有位置")
print(f"  - 复杂度从 O(n²) 降到 O(n·w)")
print(f"  - 适合长序列处理")
````

## 5. 长上下文技术 (Long Context Models)

### 5.1 长上下文的挑战

**为什么需要长上下文？**
- 文档理解（论文、书籍）
- 代码分析（整个代码库）
- 长对话历史
- 多轮推理

**Transformer 的限制**：
- 训练长度限制（通常 2K-8K tokens）
- 二次复杂度导致内存爆炸
- 位置编码外推能力差

### 5.2 扩展上下文的方法

**1. 稀疏注意力 (Sparse Attention)**

不是所有 token 都需要关注所有其他 token。

**Longformer 模式**：
- 局部窗口注意力：每个 token 关注周围 $w$ 个 token
- 全局注意力：特殊 token 关注所有位置
- 复杂度：$O(n \cdot w)$ 而非 $O(n^2)$

**BigBird 模式**：
- 局部注意力 + 全局注意力 + 随机注意力
- 理论证明：可以近似完整注意力

**2. 循环记忆 (Recurrent Memory)**

**Transformer-XL**：
- 缓存前一个段的隐藏状态
- 当前段可以关注缓存的历史
- 相对位置编码

**3. 检索增强 (Retrieval Augmentation)**

- 将长文档存储在外部数据库
- 检索相关片段
- 只对相关部分做注意力

**4. 位置编码插值 (Positional Interpolation)**

**RoPE 缩放**：
- 原始 RoPE 在训练长度外表现差
- 线性插值：将位置缩放到训练范围内
- NTK-aware 缩放：更好的外推

$$\theta_i' = \theta_i / s$$

其中 $s$ 是缩放因子。

### 5.3 长上下文模型对比

| 模型 | 方法 | 最大长度 | 复杂度 |
|------|------|---------|--------|
| **GPT-3** | 标准注意力 | 2K | $O(n^2)$ |
| **Longformer** | 稀疏注意力 | 16K | $O(n \cdot w)$ |
| **BigBird** | 混合注意力 | 16K | $O(n \cdot w)$ |
| **Transformer-XL** | 循环记忆 | 无限 | $O(n^2)$ per segment |
| **GPT-4** | 未知 | 128K | 未知 |
| **Claude 3** | 未知 | 200K | 未知 |

### 5.4 实际考虑

**训练成本**：
- 长序列训练非常昂贵
- 通常先短序列预训练，再长序列微调

**推理效率**：
- KV 缓存随长度线性增长
- 需要优化内存管理

**质量问题**：
- "中间丢失"：模型难以利用中间部分信息
- 需要特殊训练策略

````python
# 🔬 Micro Practice: Benchmark SSM vs Transformer
# Goal: Compare scaling behavior with sequence length
# Expected outcome: Demonstrate SSM's linear scaling advantage

class SimpleAttention(nn.Module):
    """Simple self-attention for comparison"""
    
    def __init__(self, d_model):
        super().__init__()
        self.d_model = d_model
        self.qkv = nn.Linear(d_model, 3 * d_model)
        self.out = nn.Linear(d_model, d_model)
    
    def forward(self, x):
        batch_size, seq_len, d_model = x.shape
        
        # Compute Q, K, V
        qkv = self.qkv(x)
        q, k, v = qkv.chunk(3, dim=-1)
        
        # Attention scores
        scores = torch.matmul(q, k.transpose(-2, -1)) / (d_model ** 0.5)
        attn = torch.softmax(scores, dim=-1)
        
        # Apply attention
        out = torch.matmul(attn, v)
        out = self.out(out)
        
        return out

# Benchmark different sequence lengths
seq_lengths = [128, 256, 512, 1024, 2048, 4096]
d_model = 256
d_state = 128

ssm_times = []
attn_times = []
ssm_memory = []
attn_memory = []

print("=" * 60)
print("SSM vs Transformer 性能对比")
print("=" * 60)
print(f"\n{'Seq Len':<10} {'SSM (ms)':<12} {'Attn (ms)':<12} {'Speedup':<10} {'Memory Ratio'}")
print("-" * 60)

for seq_len in seq_lengths:
    # Create models
    ssm_model = SimpleSSM(d_model, d_state)
    attn_model = SimpleAttention(d_model)
    
    # Create input
    x = torch.randn(1, seq_len, d_model)
    
    # Benchmark SSM
    start = time.time()
    with torch.no_grad():
        _ = ssm_model(x)
    ssm_time = (time.time() - start) * 1000
    ssm_times.append(ssm_time)
    
    # Estimate SSM memory: O(n)
    ssm_mem = seq_len * d_state
    ssm_memory.append(ssm_mem)
    
    # Benchmark Attention
    start = time.time()
    with torch.no_grad():
        _ = attn_model(x)
    attn_time = (time.time() - start) * 1000
    attn_times.append(attn_time)
    
    # Estimate Attention memory: O(n^2)
    attn_mem = seq_len * seq_len
    attn_memory.append(attn_mem)
    
    speedup = attn_time / ssm_time
    mem_ratio = attn_mem / ssm_mem
    
    print(f"{seq_len:<10} {ssm_time:<12.2f} {attn_time:<12.2f} {speedup:<10.2f}x {mem_ratio:<.1f}x")

# Visualize scaling
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Time scaling
ax1.plot(seq_lengths, ssm_times, 'o-', linewidth=2, markersize=8, 
         color='#667eea', label='SSM (O(n))')
ax1.plot(seq_lengths, attn_times, 's-', linewidth=2, markersize=8, 
         color='#f093fb', label='Attention (O(n²))')
ax1.set_xlabel('Sequence Length', fontsize=12)
ax1.set_ylabel('Time (ms)', fontsize=12)
ax1.set_title('Computation Time Scaling', fontsize=14, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)
ax1.set_xscale('log', base=2)
ax1.set_yscale('log')

# Memory scaling
ax2.plot(seq_lengths, np.array(ssm_memory) / 1e6, 'o-', linewidth=2, markersize=8, 
         color='#667eea', label='SSM (O(n))')
ax2.plot(seq_lengths, np.array(attn_memory) / 1e6, 's-', linewidth=2, markersize=8, 
         color='#f093fb', label='Attention (O(n²))')
ax2.set_xlabel('Sequence Length', fontsize=12)
ax2.set_ylabel('Memory (M elements)', fontsize=12)
ax2.set_title('Memory Usage Scaling', fontsize=14, fontweight='bold')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)
ax2.set_xscale('log', base=2)
ax2.set_yscale('log')

plt.tight_layout()
plt.show()

print("\n关键观察：")
print("  - SSM 时间复杂度：O(n) - 线性增长")
print("  - Attention 时间复杂度：O(n²) - 二次增长")
print("  - 序列越长，SSM 优势越明显")
print(f"  - 在 {seq_lengths[-1]} 长度时，SSM 快 {attn_times[-1]/ssm_times[-1]:.1f}x")
print(f"  - 内存节省：{attn_memory[-1]/ssm_memory[-1]:.1f}x")

print("\n✓ SSM 在长序列上有显著优势！")
````

````python
# 🔬 Micro Practice: Implement Simplified State Space Model
# Goal: Build basic SSM layer with linear complexity
# Expected outcome: Working SSM demonstrating O(n) scaling

class SimpleSSM(nn.Module):
    """
    Simplified State Space Model
    
    Implements discrete-time SSM:
        h_t = A * h_{t-1} + B * x_t
        y_t = C * h_t + D * x_t
    
    Args:
        d_model: Input/output dimension
        d_state: Hidden state dimension
    """
    
    def __init__(self, d_model, d_state):
        super().__init__()
        self.d_model = d_model
        self.d_state = d_state
        
        # State space parameters
        self.A = nn.Parameter(torch.randn(d_state, d_state) * 0.01)
        self.B = nn.Parameter(torch.randn(d_state, d_model) * 0.01)
        self.C = nn.Parameter(torch.randn(d_model, d_state) * 0.01)
        self.D = nn.Parameter(torch.randn(d_model, d_model) * 0.01)
        
        # Initialize A to be stable (eigenvalues < 1)
        with torch.no_grad():
            self.A.data = self.A.data * 0.9 / torch.max(torch.abs(torch.linalg.eigvals(self.A.data)))
    
    def forward(self, x):
        """
        Forward pass (sequential for clarity, can be parallelized)
        
        Args:
            x: Input tensor [batch_size, seq_len, d_model]
        
        Returns:
            y: Output tensor [batch_size, seq_len, d_model]
        """
        batch_size, seq_len, d_model = x.shape
        
        # Initialize hidden state
        h = torch.zeros(batch_size, self.d_state, device=x.device)
        
        # Process sequence
        outputs = []
        for t in range(seq_len):
            x_t = x[:, t, :]  # [batch_size, d_model]
            
            # Update state: h_t = A * h_{t-1} + B * x_t
            h = torch.matmul(h, self.A.t()) + torch.matmul(x_t, self.B.t())
            
            # Compute output: y_t = C * h_t + D * x_t
            y_t = torch.matmul(h, self.C.t()) + torch.matmul(x_t, self.D.t())
            
            outputs.append(y_t)
        
        # Stack outputs
        y = torch.stack(outputs, dim=1)  # [batch_size, seq_len, d_model]
        
        return y

# Create SSM layer
print("=" * 60)
print("State Space Model 实现")
print("=" * 60)

d_model = 512
d_state = 256

ssm = SimpleSSM(d_model, d_state)

# Test forward pass
batch_size = 4
seq_len = 1024
x = torch.randn(batch_size, seq_len, d_model)

import time

start = time.time()
output = ssm(x)
ssm_time = time.time() - start

print(f"\n输入形状: {x.shape}")
print(f"输出形状: {output.shape}")
print(f"处理时间: {ssm_time*1000:.2f} ms")

print(f"\n参数统计：")
print(f"  A 矩阵: {ssm.A.shape} = {ssm.A.numel():,} 参数")
print(f"  B 矩阵: {ssm.B.shape} = {ssm.B.numel():,} 参数")
print(f"  C 矩阵: {ssm.C.shape} = {ssm.C.numel():,} 参数")
print(f"  D 矩阵: {ssm.D.shape} = {ssm.D.numel():,} 参数")
print(f"  总计: {sum(p.numel() for p in ssm.parameters()):,} 参数")

print("\n✓ SSM 层实现完成！")
````

## 4. State Space Models (SSMs)

### 4.1 SSM 原理

**核心思想**：用线性状态空间模型替代注意力机制，实现 $O(n)$ 复杂度。

**连续时间状态空间模型**：

$$\frac{dh(t)}{dt} = Ah(t) + Bx(t)$$
$$y(t) = Ch(t) + Dx(t)$$

其中：
- $h(t)$：隐藏状态
- $x(t)$：输入
- $y(t)$：输出
- $A, B, C, D$：可学习参数矩阵

**离散化（用于实际计算）**：

$$h_t = \bar{A}h_{t-1} + \bar{B}x_t$$
$$y_t = Ch_t + Dx_t$$

其中 $\bar{A}, \bar{B}$ 是离散化后的参数。

### 4.2 SSM vs Transformer

**复杂度对比**：

| 模型 | 时间复杂度 | 空间复杂度 | 长程依赖 |
|------|-----------|-----------|---------|
| **Transformer** | $O(n^2 \cdot d)$ | $O(n^2)$ | 强 |
| **SSM** | $O(n \cdot d)$ | $O(n)$ | 强 |
| **RNN** | $O(n \cdot d^2)$ | $O(1)$ | 弱 |

**SSM 优势**：
- 线性复杂度：可处理超长序列（100K+ tokens）
- 并行训练：可以并行计算（类似 Transformer）
- 高效推理：递归形式，常数时间生成
- 长程建模：理论上可以捕获任意长的依赖

### 4.3 S4 和 Mamba

**S4 (Structured State Space)**：
- 使用特殊的结构化矩阵（HiPPO）
- 解决了训练不稳定问题
- 在长程任务上超越 Transformer

**Mamba (2023)**：
- 选择性状态空间模型
- 输入依赖的参数（$A, B, C$ 依赖输入）
- 在语言建模上接近 Transformer 性能
- 推理速度快 5-10x

### 4.4 SSM 的挑战

**1. 内容依赖的注意力**：
- SSM 是线性的，难以做内容依赖的选择
- Transformer 的注意力可以动态关注不同位置
- Mamba 通过选择性机制部分解决

**2. 训练技巧**：
- 需要特殊的初始化
- 离散化步长需要调优
- 数值稳定性问题

**3. 生态系统**：
- 工具和库不如 Transformer 成熟
- 预训练模型较少
- 社区支持有限

````python
# 🔬 Micro Practice: Compare MoE vs Dense Models
# Goal: Analyze parameter count and computation trade-offs
# Expected outcome: Clear comparison showing MoE efficiency

def count_parameters(model):
    """Count trainable parameters"""
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

def estimate_flops(model, input_shape):
    """Estimate FLOPs for forward pass"""
    # Simplified estimation
    if isinstance(model, MixtureOfExperts):
        # MoE: only k experts are active
        expert_params = count_parameters(model.experts[0])
        gate_params = count_parameters(model.gate)
        # FLOPs = gate computation + k * expert computation
        batch_size, seq_len, d_model = input_shape
        flops = batch_size * seq_len * (gate_params + model.k * expert_params)
    else:
        # Dense: all parameters are active
        params = count_parameters(model)
        batch_size, seq_len, d_model = input_shape
        flops = batch_size * seq_len * params
    return flops

# Create models for comparison
d_model = 512
d_ff = 2048
num_experts = 8
k = 2

# MoE model
moe_model = MixtureOfExperts(num_experts, d_model, d_ff, k)

# Dense model (equivalent to single expert)
dense_model = Expert(d_model, d_ff)

# Count parameters
moe_params = count_parameters(moe_model)
dense_params = count_parameters(dense_model)

print("=" * 60)
print("MoE vs Dense 模型对比")
print("=" * 60)

print(f"\n参数量对比：")
print(f"  Dense 模型: {dense_params:,} 参数")
print(f"  MoE 模型:   {moe_params:,} 参数")
print(f"  参数比例:   {moe_params/dense_params:.1f}x")

print(f"\n计算量对比（每个 token）：")
print(f"  Dense 模型: 使用 100% 参数")
print(f"  MoE 模型:   使用 {k/num_experts*100:.1f}% 参数 ({k}/{num_experts} 专家)")
print(f"  计算比例:   ~{k/num_experts:.2f}x of dense")

print(f"\n效率分析：")
print(f"  参数效率: {(moe_params/dense_params) / (k/num_experts):.1f}x")
print(f"  说明: MoE 用 {k/num_experts:.1%} 的计算获得 {moe_params/dense_params:.1f}x 的参数")

# Visualize comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Parameters comparison
models = ['Dense', 'MoE']
params = [dense_params, moe_params]
colors = ['#667eea', '#f093fb']

ax1.bar(models, params, color=colors, alpha=0.7)
ax1.set_ylabel('Parameters', fontsize=12)
ax1.set_title('Parameter Count Comparison', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')

for i, (model, param) in enumerate(zip(models, params)):
    ax1.text(i, param, f'{param/1e6:.1f}M', ha='center', va='bottom', fontsize=11)

# Computation comparison
computation = [1.0, k/num_experts]

ax2.bar(models, computation, color=colors, alpha=0.7)
ax2.set_ylabel('Relative Computation', fontsize=12)
ax2.set_title('Computation Cost Comparison', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')
ax2.set_ylim([0, 1.2])

for i, (model, comp) in enumerate(zip(models, computation)):
    ax2.text(i, comp, f'{comp:.1%}', ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()

print("\n✓ MoE 效率优势明显！")
print("\n关键洞察：")
print("  - MoE 可以用更少的计算获得更多的参数")
print("  - 适合需要大容量但计算受限的场景")
print("  - 训练和推理都能受益")
````

### 3.4 MoE vs Dense 对比

让我们比较 MoE 和密集模型的效率。

````python
# 🔬 Micro Practice: Implement Mixture-of-Experts Layer
# Goal: Build MoE with top-k gating and load balancing
# Expected outcome: Working MoE layer with expert routing

class Expert(nn.Module):
    """Single expert network (FFN)"""
    
    def __init__(self, d_model, d_ff):
        super().__init__()
        self.fc1 = nn.Linear(d_model, d_ff)
        self.fc2 = nn.Linear(d_ff, d_model)
        self.activation = nn.ReLU()
    
    def forward(self, x):
        return self.fc2(self.activation(self.fc1(x)))

class MixtureOfExperts(nn.Module):
    """
    Mixture-of-Experts Layer
    
    Args:
        num_experts: Number of expert networks
        d_model: Model dimension
        d_ff: FFN hidden dimension
        k: Number of experts to activate (top-k)
    """
    
    def __init__(self, num_experts, d_model, d_ff, k=2):
        super().__init__()
        self.num_experts = num_experts
        self.k = k
        self.d_model = d_model
        
        # Create experts
        self.experts = nn.ModuleList([
            Expert(d_model, d_ff) for _ in range(num_experts)
        ])
        
        # Gating network
        self.gate = nn.Linear(d_model, num_experts)
        
        # For tracking expert usage
        self.expert_usage = torch.zeros(num_experts)
    
    def forward(self, x):
        """
        Forward pass with top-k gating
        
        Args:
            x: Input tensor [batch_size, seq_len, d_model]
        
        Returns:
            output: MoE output [batch_size, seq_len, d_model]
            load_balance_loss: Load balancing loss
        """
        batch_size, seq_len, d_model = x.shape
        
        # Flatten for gating
        x_flat = x.view(-1, d_model)  # [batch_size * seq_len, d_model]
        
        # Compute gate scores
        gate_logits = self.gate(x_flat)  # [batch_size * seq_len, num_experts]
        
        # Top-k gating
        top_k_logits, top_k_indices = torch.topk(gate_logits, self.k, dim=-1)
        top_k_gates = torch.softmax(top_k_logits, dim=-1)
        
        # Initialize output
        output = torch.zeros_like(x_flat)
        
        # Track expert usage for load balancing
        expert_mask = torch.zeros(x_flat.size(0), self.num_experts, device=x.device)
        
        # Route to experts
        for i in range(self.k):
            expert_idx = top_k_indices[:, i]  # [batch_size * seq_len]
            gate_weight = top_k_gates[:, i:i+1]  # [batch_size * seq_len, 1]
            
            # Update expert usage mask
            expert_mask.scatter_(1, expert_idx.unsqueeze(1), 1.0)
            
            # Process each expert
            for expert_id in range(self.num_experts):
                # Find tokens routed to this expert
                mask = (expert_idx == expert_id)
                if mask.any():
                    # Get tokens for this expert
                    expert_input = x_flat[mask]
                    
                    # Process through expert
                    expert_output = self.experts[expert_id](expert_input)
                    
                    # Add weighted output
                    output[mask] += gate_weight[mask] * expert_output
                    
                    # Track usage
                    self.expert_usage[expert_id] += mask.sum().item()
        
        # Reshape output
        output = output.view(batch_size, seq_len, d_model)
        
        # Compute load balance loss
        # Encourage uniform expert usage
        expert_usage_ratio = expert_mask.mean(0)  # [num_experts]
        target_ratio = 1.0 / self.num_experts
        load_balance_loss = torch.sum((expert_usage_ratio - target_ratio) ** 2)
        
        return output, load_balance_loss

# Create MoE layer
print("=" * 60)
print("Mixture-of-Experts 实现")
print("=" * 60)

num_experts = 8
d_model = 512
d_ff = 2048
k = 2

moe = MixtureOfExperts(num_experts, d_model, d_ff, k)

# Test forward pass
batch_size = 4
seq_len = 16
x = torch.randn(batch_size, seq_len, d_model)

output, load_balance_loss = moe(x)

print(f"\n输入形状: {x.shape}")
print(f"输出形状: {output.shape}")
print(f"负载均衡损失: {load_balance_loss.item():.4f}")

print(f"\n专家使用统计：")
print(f"  总专家数: {num_experts}")
print(f"  每次激活: {k} 个专家")
print(f"  理论计算量: {k}/{num_experts} = {k/num_experts:.1%} of dense model")

# Visualize expert usage
expert_usage_normalized = moe.expert_usage / moe.expert_usage.sum()

plt.figure(figsize=(10, 5))
plt.bar(range(num_experts), expert_usage_normalized.numpy(), color='#667eea', alpha=0.7)
plt.axhline(y=1/num_experts, color='red', linestyle='--', label='Uniform distribution')
plt.xlabel('Expert ID', fontsize=12)
plt.ylabel('Usage Ratio', fontsize=12)
plt.title('Expert Usage Distribution', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print("\n✓ MoE 层实现完成！")
print("\n关键特性：")
print("  - Top-k 门控（稀疏激活）")
print("  - 负载均衡损失")
print("  - 专家使用追踪")
print("  - 参数高效扩展")
````

## 3. Mixture-of-Experts (MoE)

### 3.1 MoE 原理

**核心思想**：稀疏激活 - 每次只使用部分专家网络。

**架构**：

$$y = \sum_{i=1}^N G(x)_i \cdot E_i(x)$$

其中：
- $G(x)$：门控网络（Gating Network），输出专家权重
- $E_i(x)$：第 $i$ 个专家网络（Expert Network）
- $N$：专家总数

**Top-K 门控**：

$$G(x) = \text{Softmax}(\text{TopK}(x \cdot W_g, k))$$

只激活得分最高的 $k$ 个专家（通常 $k=1$ 或 $k=2$）。

### 3.2 MoE 优势

**1. 参数扩展不增加计算**：
- 密集模型：1B 参数 → 1B FLOPs
- MoE 模型：10B 参数 → 1B FLOPs（10 个专家，激活 1 个）

**2. 专家专业化**：
- 不同专家学习不同模式
- 自动任务分解
- 更好的泛化能力

**3. 实际应用**：
- Switch Transformer (Google, 2021)：1.6T 参数
- GLaM (Google, 2021)：1.2T 参数
- GPT-4 (传闻)：使用 MoE

### 3.3 MoE 挑战

**1. 负载均衡**：
- 问题：某些专家过度使用，其他专家闲置
- 解决：添加负载均衡损失

$$L_{\text{balance}} = \alpha \cdot \text{CV}(\text{expert\_usage})^2$$

**2. 训练不稳定**：
- 门控网络难以训练
- 专家崩溃（collapse）
- 需要特殊初始化

**3. 通信开销**：
- 分布式训练时专家分布在不同设备
- All-to-All 通信成本高
- 需要优化通信策略

- 位置编码限制
- 训练-推理不匹配
- 外推能力差

**3. 计算效率**：
- 密集激活（所有参数都参与）
- 无法动态调整计算量
- 推理成本高

**4. 单模态限制**：
- 仅处理文本
- 缺乏视觉理解
- 多模态需要额外设计

### 📊 缩放定律 (Scaling Laws)

**Kaplan et al. (2020) 发现**：

$$L(N) = \left(\frac{N_c}{N}\right)^{\alpha_N}$$

其中：
- $L$：损失
- $N$：参数量
- $N_c, \alpha_N$：常数

**关键发现**：
- 模型越大，性能越好（幂律关系）
- 数据和计算同样重要
- 存在最优的参数-数据-计算比例

**Chinchilla 优化 (2022)**：
- 之前：过度参数化，训练不足
- 优化：参数和数据量应同步增长
- 结果：70B Chinchilla > 280B Gopher

## 2. Transformer 复杂度分析 (Complexity Analysis)

### 2.1 注意力机制复杂度

**标准自注意力**：

$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$

**复杂度分析**：
- $QK^T$：$O(n^2 \cdot d)$ - 序列长度的平方
- Softmax：$O(n^2)$
- 乘以 $V$：$O(n^2 \cdot d)$
- **总计**：$O(n^2 \cdot d)$ 时间，$O(n^2)$ 内存

### 2.2 实际影响

让我们可视化不同序列长度下的计算和内存需求。

````python
# 🔬 Micro Practice: Visualize Transformer Complexity
# Goal: Demonstrate quadratic scaling of attention
# Expected outcome: Clear visualization of complexity growth

def compute_attention_complexity(seq_lengths, d_model=512):
    """
    Compute attention complexity for different sequence lengths
    
    Args:
        seq_lengths: List of sequence lengths
        d_model: Model dimension
    
    Returns:
        flops: FLOPs for each sequence length
        memory: Memory (in elements) for each sequence length
    """
    flops = []
    memory = []
    
    for n in seq_lengths:
        # FLOPs: QK^T (n^2 * d) + softmax (n^2) + matmul with V (n^2 * d)
        flop = 2 * n * n * d_model + n * n + n * n * d_model
        flops.append(flop)
        
        # Memory: attention matrix (n^2)
        mem = n * n
        memory.append(mem)
    
    return np.array(flops), np.array(memory)

# Sequence lengths from 128 to 32K
seq_lengths = [128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768]
flops, memory = compute_attention_complexity(seq_lengths)

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# FLOPs
ax1.plot(seq_lengths, flops / 1e9, 'o-', linewidth=2, markersize=8, color='#667eea')
ax1.set_xlabel('Sequence Length', fontsize=12)
ax1.set_ylabel('GFLOPs', fontsize=12)
ax1.set_title('Attention Computation (FLOPs)', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.set_xscale('log', base=2)
ax1.set_yscale('log')

# Add annotations
for i, (x, y) in enumerate(zip(seq_lengths[::2], flops[::2] / 1e9)):
    ax1.annotate(f'{x}', xy=(x, y), xytext=(5, 5), 
                textcoords='offset points', fontsize=9)

# Memory
ax2.plot(seq_lengths, memory / 1e6, 's-', linewidth=2, markersize=8, color='#f093fb')
ax2.set_xlabel('Sequence Length', fontsize=12)
ax2.set_ylabel('Memory (M elements)', fontsize=12)
ax2.set_title('Attention Memory Usage', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_xscale('log', base=2)
ax2.set_yscale('log')

# Add annotations
for i, (x, y) in enumerate(zip(seq_lengths[::2], memory[::2] / 1e6)):
    ax2.annotate(f'{x}', xy=(x, y), xytext=(5, 5), 
                textcoords='offset points', fontsize=9)

plt.tight_layout()
plt.show()

print("=" * 60)
print("Transformer 复杂度分析")
print("=" * 60)
print("\n序列长度对比：\n")
print(f"{'Length':<10} {'FLOPs (G)':<15} {'Memory (MB)':<15} {'相对增长'}")
print("-" * 60)

base_flops = flops[0]
base_memory = memory[0]

for i, n in enumerate(seq_lengths):
    flop_ratio = flops[i] / base_flops
    mem_ratio = memory[i] / base_memory
    print(f"{n:<10} {flops[i]/1e9:<15.2f} {memory[i]*4/1e6:<15.2f} {flop_ratio:.1f}x")

print("\n关键观察：")
print("  - 序列长度翻倍 → 计算量增加 4 倍")
print("  - 序列长度翻倍 → 内存增加 4 倍")
print("  - 32K 序列比 128 序列慢 64K 倍！")
print("\n这就是为什么需要新架构！")
````

### 问题 1：MoE（混合专家模型）的负载均衡问题：如果门控网络总是选择同一批专家，会产生什么问题？应如何解决？

**问题根源**：MoE 门控网络若总选择同一批专家，会导致**专家崩溃（Expert Collapse）**现象：

- 被频繁选中的专家持续接收梯度更新，能力越来越强
- 未被选中的专家缺乏梯度信号，参数几乎不更新，形同"僵尸专家"
- 模型实际上退化为 Dense 模型，失去 MoE 的参数效率优势

**理论分析**：设专家被选中的概率为 $p_i$，若 $p_i \to 1$ 对某个 $i$ 成立，则负载均衡损失：

$$\mathcal{L}_{\text{balance}} = \alpha \cdot N \cdot \sum_{i=1}^{N} f_i \cdot P_i$$

其中 $f_i$ 为专家 $i$ 被选中的比例，$P_i$ 为门控分数均值。完全崩溃时该损失极大，但若无该约束则无法自我纠正。

**解决方案**：

| 方法 | 原理 | 效果 |
|------|------|------|
| 辅助负载均衡损失 | 惩罚专家选择分布的不均匀性 | 效果好，是主流方案 |
| 专家容量上限 | 每个专家最多处理 $C$ 个 token | 强制均衡，但可能丢弃 token |
| Expert Choice | 改为专家主动选择 token | 天然均衡，但需通信协调 |
| 随机路由抖动 | 门控加入噪声 | 防止早期崩溃，需配合其他方法 |

**推荐实践**：Switch Transformer 使用辅助损失 + 容量因子（capacity factor）$c=1.25$，允许每个专家处理 $1.25 \times \frac{T}{N}$ 个 token（$T$ 为总 token 数，$N$ 为专家数），兼顾均衡与灵活性。

### 问题 2：SSM（状态空间模型）相比 Transformer 的核心优势是线性复杂度，但它的弱点是什么？在哪些任务上 Transformer 仍然更优？

**SSM 的复杂度优势**：

- Transformer 自注意力：时间 $O(L^2 d)$，内存 $O(L^2)$
- SSM（如 Mamba）：时间 $O(Ld_s)$，内存 $O(Ld_s)$，其中 $L$ 为序列长度，$d_s$ 为状态维度

**SSM 的核心弱点**：

1. **内容依赖检索能力弱**：SSM 通过固定大小的隐藏状态压缩历史信息，对"精确内容匹配"能力较弱（例如：在 100K token 中找到某个特定实体）。Transformer 注意力天然支持任意位置的精确 Key-Value 查找。

2. **上下文窗口的"软遗忘"**：SSM 的递归结构导致远期信息被指数级衰减，而 Transformer 理论上可以对任意位置施加等权注意力。

3. **训练并行度受限于 selective scan**：虽然 Mamba 的 selective scan 已经是 $O(L)$，但相比 Transformer 的矩阵运算，GPU 利用率仍有差距。

**Transformer 仍占优的任务**：

| 任务类型 | 原因 |
|---------|------|
| **In-Context Learning（ICL）** | 需要精确关注 prompt 中的示例，注意力的精确检索能力更强 |
| **信息检索型问答** | 需要从长文档中精确定位某个答案 |
| **代码生成** | 频繁需要"复制"远距离的变量名、函数签名等精确 token |
| **关联推理** | 如"A 比 B 大，B 比 C 大，A 和 C 的关系？"需要显式关系建模 |

**结论**：混合架构（如 Jamba）是当前趋势——用 SSM 层处理长程序列建模，用 Transformer 层处理内容依赖的精确检索，兼顾效率与能力。

### 问题 3：RoPE（旋转位置编码）相比绝对位置编码有何优势？位置插值（Position Interpolation）如何帮助 LLM 处理超出训练长度的序列？

**RoPE 的设计原理**：

RoPE 将位置信息编码为旋转矩阵，使得 token $m$ 和 $n$ 的注意力分数只依赖于**相对距离** $m-n$：

$$\langle f_q(\mathbf{x}_m, m), f_k(\mathbf{x}_n, n) \rangle = g(\mathbf{x}_m, \mathbf{x}_n, m-n)$$

**相比绝对位置编码的优势**：

| 特性 | 绝对位置编码（APE） | RoPE |
|------|---------------------|------|
| 相对位置感知 | 间接（需学习） | 天然支持 |
| 外推能力 | 差（训练外位置未见过） | 较好（旋转角度连续） |
| 序列长度灵活性 | 固定 | 可扩展 |
| 理论基础 | 启发式 | 数学严格 |

**位置插值（PI）的原理**：

设训练长度为 $L$，目标长度为 $L'$（$L' > L$）。朴素外推会遇到旋转角度超出训练分布的问题。位置插值将位置 $m$ 缩放为 $m \cdot \frac{L}{L'}$：

$$\text{RoPE-PI}(m) = \text{RoPE}\left(\frac{m \cdot L}{L'}\right)$$

**效果**：将每个位置"压缩"到训练范围内，旋转角度不会超出训练分布，只需少量微调（~1000 步）即可适配新长度。

**局限性**：过度压缩会导致近距离 token 的位置区分度下降（因为相邻位置的角度差变小），影响局部语义建模。YaRN 等改进方法通过非均匀缩放解决了这一问题。

### 问题 4：多模态大模型（如 LLaVA）中，视觉特征与语言特征是如何对齐的？训练时有哪些关键挑战？

**视觉-语言对齐的核心机制**：

```
图像 → 视觉编码器（ViT/CLIP）→ 视觉 token → 投影层 → 语言 token 空间 → LLM
```

主流方法（以 LLaVA 为例）：
1. **视觉编码器**：使用预训练的 CLIP ViT 提取图像特征，输出固定数量的视觉 token（如 $14 \times 14 = 196$ 个 patch token）
2. **投影层**（MLP Connector）：将视觉特征维度投影到语言模型的嵌入维度
3. **联合建模**：视觉 token 与文本 token 拼接后，统一输入 LLM 进行自回归生成

**训练阶段的关键挑战**：

| 挑战 | 描述 | 解决方案 |
|------|------|---------|
| **特征空间不对齐** | 视觉特征分布与语言嵌入分布差异大 | 两阶段训练：先冻结 LLM 预训练投影层，再端到端微调 |
| **灾难性遗忘** | 多模态训练可能损害纯文本能力 | 混入文本数据；使用 LoRA 限制更新范围 |
| **幻觉问题** | 模型生成与图像不符的描述 | 引入负样本对比训练；视觉接地（Visual Grounding）训练 |
| **分辨率限制** | 固定 patch 数量对高分辨率图像丢失细节 | 动态分辨率（如 LLaVA-HD）：将图像切片后分别编码 |
| **数据规模** | 高质量图文对数据稀缺 | 使用 GPT-4V 生成的合成数据（如 LLaVA-Instruct-150K）|

**关键洞察**：多模态对齐的难点不在于技术复杂度，而在于**如何设计高质量的对齐数据**。LLaVA 的成功很大程度上来自于使用 GPT-4 生成的多样化指令微调数据。

### 问题 5：GQA（Grouped Query Attention）是 MQA 和 MHA 的折中方案，请分析三者的内存效率和质量的权衡关系，并说明为何现代 LLM 普遍采用 GQA？

**三种注意力机制对比**：

设 $H$ 为注意力头数，$G$ 为 GQA 的分组数（$1 \leq G \leq H$）：

| 机制 | KV 头数 | KV Cache 大小 | 质量 | 适用场景 |
|------|---------|--------------|------|---------|
| **MHA**（Multi-Head Attention） | $H$ | $2 \times H \times d_k \times L$ | 最高 | 训练为主，推理资源充足 |
| **MQA**（Multi-Query Attention） | $1$ | $2 \times 1 \times d_k \times L$ | 较低（尤其长文本） | 极度资源受限 |
| **GQA**（Grouped Query Attention） | $G$ | $2 \times G \times d_k \times L$ | 接近 MHA | 平衡效率与质量 |

**数学说明**：

MHA 注意力计算：每个 Query 头 $i$ 独立计算 $\text{Attn}(Q_i, K_i, V_i)$

GQA 注意力计算：每组 $g$ 个 Query 头共享同一组 KV，组内计算 $\text{Attn}(Q_{g,j}, K_g, V_g)$

**GQA 被普遍采用的原因**：

1. **推理内存瓶颈**：LLM 推理时，KV Cache 是主要内存开销。对于 LLaMA-70B（$H=64$），MHA 的 KV Cache 约为 GQA（$G=8$）的 8 倍，意味着在同等显存下 GQA 能支持更长上下文或更大 batch size。

2. **质量-效率帕累托最优**：实验表明 GQA 在适当分组数下（通常 $G = H/8$ 到 $H/4$）性能接近 MHA，远优于 MQA，而内存开销仅为 MHA 的 $G/H$。

3. **无需重头训练**：现有 MHA 模型可通过 "Mean Pooling" 将相邻 KV 头合并，用少量微调转换为 GQA（Llama 2 使用的方法）。

**结论**：GQA 是推理优化的工程最优解，LLaMA 2/3、Mistral、Gemma 等主流模型均采用 GQA，成为现代 LLM 的标准组件。
