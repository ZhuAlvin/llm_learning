````python
# 🔬 Micro Practice: Rich Message Renderer
# Goal: Implement markdown, code highlighting, and math rendering
# Expected outcome: Production-ready message renderer

rich_renderer_code = '''
// ============================================
// Rich Message Renderer with Markdown, Code, and Math
// ============================================
import React from 'react';
import ReactMarkdown from 'react-markdown';
import { Prism as SyntaxHighlighter } from 'react-syntax-highlighter';
import { vscDarkPlus } from 'react-syntax-highlighter/dist/esm/styles/prism';
import remarkMath from 'remark-math';
import rehypeKatex from 'rehype-katex';
import 'katex/dist/katex.min.css';

interface RichMessageProps {
  content: string;
  role: 'user' | 'assistant';
}

const RichMessage: React.FC<RichMessageProps> = ({ content, role }) => {
  const [copied, setCopied] = React.useState(false);
  
  const handleCopy = (code: string) => {
    navigator.clipboard.writeText(code);
    setCopied(true);
    setTimeout(() => setCopied(false), 2000);
  };
  
  return (
    <div className={`rich-message ${role}`}>
      <ReactMarkdown
        remarkPlugins={[remarkMath]}
        rehypePlugins={[rehypeKatex]}
        components={{
          // Code block with syntax highlighting
          code({ node, inline, className, children, ...props }) {
            const match = /language-(\\w+)/.exec(className || '');
            const code = String(children).replace(/\\n$/, '');
            
            return !inline && match ? (
              <div className="code-block">
                <div className="code-header">
                  <span className="language">{match[1]}</span>
                  <button 
                    onClick={() => handleCopy(code)}
                    className="copy-button"
                  >
                    {copied ? '✓ Copied' : '📋 Copy'}
                  </button>
                </div>
                <SyntaxHighlighter
                  style={vscDarkPlus}
                  language={match[1]}
                  PreTag="div"
                  showLineNumbers
                  {...props}
                >
                  {code}
                </SyntaxHighlighter>
              </div>
            ) : (
              <code className="inline-code" {...props}>
                {children}
              </code>
            );
          },
          
          // Links with security
          a({ node, children, href, ...props }) {
            return (
              <a 
                href={href} 
                target="_blank" 
                rel="noopener noreferrer"
                {...props}
              >
                {children}
              </a>
            );
          },
          
          // Images with lazy loading
          img({ node, src, alt, ...props }) {
            return (
              <img 
                src={src} 
                alt={alt} 
                loading="lazy"
                className="message-image"
                {...props}
              />
            );
          },
          
          // Tables
          table({ node, children, ...props }) {
            return (
              <div className="table-wrapper">
                <table {...props}>{children}</table>
              </div>
            );
          }
        }}
      >
        {content}
      </ReactMarkdown>
    </div>
  );
};

// ============================================
// Message Actions Component
// ============================================
interface MessageActionsProps {
  messageId: string;
  content: string;
  onEdit: (id: string) => void;
  onRegenerate: (id: string) => void;
  onDelete: (id: string) => void;
}

const MessageActions: React.FC<MessageActionsProps> = ({
  messageId,
  content,
  onEdit,
  onRegenerate,
  onDelete
}) => {
  const [showActions, setShowActions] = React.useState(false);
  
  const handleCopy = () => {
    navigator.clipboard.writeText(content);
  };
  
  return (
    <div 
      className="message-actions"
      onMouseEnter={() => setShowActions(true)}
      onMouseLeave={() => setShowActions(false)}
    >
      {showActions && (
        <div className="action-buttons">
          <button onClick={handleCopy} title="Copy">
            📋
          </button>
          <button onClick={() => onEdit(messageId)} title="Edit">
            ✏️
          </button>
          <button onClick={() => onRegenerate(messageId)} title="Regenerate">
            🔄
          </button>
          <button onClick={() => onDelete(messageId)} title="Delete">
            🗑️
          </button>
        </div>
      )}
    </div>
  );
};

// ============================================
// Export Conversation Component
// ============================================
interface ExportOptions {
  format: 'json' | 'markdown' | 'text';
  includeTimestamps: boolean;
}

const ExportConversation: React.FC<{ messages: Message[] }> = ({ messages }) => {
  const exportToJSON = () => {
    const data = JSON.stringify(messages, null, 2);
    downloadFile(data, 'conversation.json', 'application/json');
  };
  
  const exportToMarkdown = () => {
    let markdown = '# Conversation\\n\\n';
    messages.forEach(msg => {
      markdown += `## ${msg.role === 'user' ? 'User' : 'Assistant'}\\n\\n`;
      markdown += `${msg.content}\\n\\n`;
      markdown += `*${msg.timestamp.toLocaleString()}*\\n\\n---\\n\\n`;
    });
    downloadFile(markdown, 'conversation.md', 'text/markdown');
  };
  
  const exportToText = () => {
    let text = '';
    messages.forEach(msg => {
      text += `[${msg.role.toUpperCase()}] ${msg.timestamp.toLocaleString()}\\n`;
      text += `${msg.content}\\n\\n`;
    });
    downloadFile(text, 'conversation.txt', 'text/plain');
  };
  
  const downloadFile = (content: string, filename: string, type: string) => {
    const blob = new Blob([content], { type });
    const url = URL.createObjectURL(blob);
    const a = document.createElement('a');
    a.href = url;
    a.download = filename;
    a.click();
    URL.revokeObjectURL(url);
  };
  
  return (
    <div className="export-menu">
      <button onClick={exportToJSON}>Export as JSON</button>
      <button onClick={exportToMarkdown}>Export as Markdown</button>
      <button onClick={exportToText}>Export as Text</button>
    </div>
  );
};

export { RichMessage, MessageActions, ExportConversation };
'''

print("=" * 60)
print("富文本消息渲染器")
print("=" * 60)
print("\n代码示例：\n")
print(rich_renderer_code)

print("\n" + "=" * 60)
print("依赖安装：")
print("=" * 60)
print("\nnpm install react-markdown remark-math rehype-katex")
print("npm install react-syntax-highlighter")
print("npm install @types/react-syntax-highlighter")
print("npm install katex")

print("\n✓ 富文本渲染器代码已生成！")
print("\n关键特性：")
print("  - Markdown 渲染（react-markdown）")
print("  - 代码语法高亮（Prism）")
print("  - LaTeX 数学公式（KaTeX）")
print("  - 复制代码按钮")
print("  - 消息操作（编辑、重新生成、删除）")
print("  - 导出对话（JSON, Markdown, Text）")
print("  - 图片懒加载")
print("  - 安全的外部链接")
````

````python
# 🔬 Micro Practice: React Chat Components
# Goal: Build React components for LLM chat
# Expected outcome: Production-ready React component structure

# Note: This demonstrates React/TypeScript structure
# To use: Create React app with `npx create-react-app my-chat --template typescript`

react_components = '''
// ============================================
// 1. Message Component
// ============================================
import React from 'react';

interface MessageProps {
  role: 'user' | 'assistant';
  content: string;
  timestamp: Date;
}

const Message: React.FC<MessageProps> = ({ role, content, timestamp }) => {
  return (
    <div className={`message ${role}`}>
      <div className="message-content">
        {content}
      </div>
      <div className="message-time">
        {timestamp.toLocaleTimeString()}
      </div>
    </div>
  );
};

// ============================================
// 2. MessageList Component
// ============================================
import React, { useEffect, useRef } from 'react';

interface Message {
  id: string;
  role: 'user' | 'assistant';
  content: string;
  timestamp: Date;
}

interface MessageListProps {
  messages: Message[];
}

const MessageList: React.FC<MessageListProps> = ({ messages }) => {
  const messagesEndRef = useRef<HTMLDivElement>(null);
  
  // Auto-scroll to bottom
  useEffect(() => {
    messagesEndRef.current?.scrollIntoView({ behavior: 'smooth' });
  }, [messages]);
  
  return (
    <div className="message-list">
      {messages.map((message) => (
        <Message
          key={message.id}
          role={message.role}
          content={message.content}
          timestamp={message.timestamp}
        />
      ))}
      <div ref={messagesEndRef} />
    </div>
  );
};

// ============================================
// 3. ChatInput Component
// ============================================
import React, { useState } from 'react';

interface ChatInputProps {
  onSend: (message: string) => void;
  disabled?: boolean;
}

const ChatInput: React.FC<ChatInputProps> = ({ onSend, disabled = false }) => {
  const [input, setInput] = useState('');
  
  const handleSubmit = (e: React.FormEvent) => {
    e.preventDefault();
    if (input.trim() && !disabled) {
      onSend(input);
      setInput('');
    }
  };
  
  return (
    <form onSubmit={handleSubmit} className="chat-input">
      <input
        type="text"
        value={input}
        onChange={(e) => setInput(e.target.value)}
        placeholder="Type your message..."
        disabled={disabled}
      />
      <button type="submit" disabled={disabled || !input.trim()}>
        Send
      </button>
    </form>
  );
};

// ============================================
// 4. ChatContainer Component (Main)
// ============================================
import React, { useState } from 'react';

const ChatContainer: React.FC = () => {
  const [messages, setMessages] = useState<Message[]>([]);
  const [isLoading, setIsLoading] = useState(false);
  
  const handleSendMessage = async (content: string) => {
    // Add user message
    const userMessage: Message = {
      id: Date.now().toString(),
      role: 'user',
      content,
      timestamp: new Date()
    };
    
    setMessages(prev => [...prev, userMessage]);
    setIsLoading(true);
    
    try {
      // Call API with streaming
      const response = await fetch('/api/chat', {
        method: 'POST',
        headers: { 'Content-Type': 'application/json' },
        body: JSON.stringify({ message: content })
      });
      
      const reader = response.body?.getReader();
      const decoder = new TextDecoder();
      
      let assistantMessage: Message = {
        id: (Date.now() + 1).toString(),
        role: 'assistant',
        content: '',
        timestamp: new Date()
      };
      
      setMessages(prev => [...prev, assistantMessage]);
      
      // Stream response
      while (true) {
        const { done, value } = await reader!.read();
        if (done) break;
        
        const chunk = decoder.decode(value);
        assistantMessage.content += chunk;
        
        setMessages(prev => 
          prev.map(m => m.id === assistantMessage.id ? assistantMessage : m)
        );
      }
    } catch (error) {
      console.error('Error:', error);
    } finally {
      setIsLoading(false);
    }
  };
  
  return (
    <div className="chat-container">
      <MessageList messages={messages} />
      <ChatInput onSend={handleSendMessage} disabled={isLoading} />
    </div>
  );
};

export default ChatContainer;

// ============================================
// 5. Custom Hook for Streaming
// ============================================
import { useState, useCallback } from 'react';

interface UseStreamingChatReturn {
  messages: Message[];
  sendMessage: (content: string) => Promise<void>;
  isLoading: boolean;
}

const useStreamingChat = (apiUrl: string): UseStreamingChatReturn => {
  const [messages, setMessages] = useState<Message[]>([]);
  const [isLoading, setIsLoading] = useState(false);
  
  const sendMessage = useCallback(async (content: string) => {
    const userMessage: Message = {
      id: Date.now().toString(),
      role: 'user',
      content,
      timestamp: new Date()
    };
    
    setMessages(prev => [...prev, userMessage]);
    setIsLoading(true);
    
    try {
      const eventSource = new EventSource(`${apiUrl}?message=${encodeURIComponent(content)}`);
      
      let assistantContent = '';
      const assistantMessage: Message = {
        id: (Date.now() + 1).toString(),
        role: 'assistant',
        content: '',
        timestamp: new Date()
      };
      
      setMessages(prev => [...prev, assistantMessage]);
      
      eventSource.onmessage = (event) => {
        const data = JSON.parse(event.data);
        assistantContent += data.token;
        
        setMessages(prev =>
          prev.map(m => 
            m.id === assistantMessage.id 
              ? { ...m, content: assistantContent }
              : m
          )
        );
      };
      
      eventSource.onerror = () => {
        eventSource.close();
        setIsLoading(false);
      };
    } catch (error) {
      console.error('Error:', error);
      setIsLoading(false);
    }
  }, [apiUrl]);
  
  return { messages, sendMessage, isLoading };
};
'''

print("=" * 60)
print("React 聊天组件示例")
print("=" * 60)
print("\n组件结构：\n")
print(react_components)

print("\n" + "=" * 60)
print("项目设置：")
print("=" * 60)
print("\n1. 创建 React 应用:")
print("   npx create-react-app my-chat --template typescript")
print("\n2. 安装依赖:")
print("   npm install axios react-markdown")
print("\n3. 复制组件代码到 src/components/")
print("\n4. 添加 CSS 样式")
print("\n5. 运行开发服务器:")
print("   npm start")

print("\n✓ React 组件代码已生成！")
print("\n关键特性：")
print("  - TypeScript 类型安全")
print("  - 组件化设计")
print("  - 自定义 Hook (useStreamingChat)")
print("  - 流式响应处理")
print("  - 自动滚动")
print("  - 状态管理")
````

````python
# 🔬 Micro Practice: WebSocket Backend
# Goal: Implement FastAPI WebSocket endpoint
# Expected outcome: Real-time bidirectional chat

from fastapi import WebSocket, WebSocketDisconnect
from typing import List, Dict
import time

class ConnectionManager:
    """Manage WebSocket connections"""
    
    def __init__(self):
        self.active_connections: List[WebSocket] = []
        self.message_history: List[Dict] = []
    
    async def connect(self, websocket: WebSocket):
        """Accept new connection"""
        await websocket.accept()
        self.active_connections.append(websocket)
        print(f"Client connected. Total connections: {len(self.active_connections)}")
    
    def disconnect(self, websocket: WebSocket):
        """Remove connection"""
        self.active_connections.remove(websocket)
        print(f"Client disconnected. Total connections: {len(self.active_connections)}")
    
    async def send_personal_message(self, message: str, websocket: WebSocket):
        """Send message to specific client"""
        await websocket.send_text(message)
    
    async def broadcast(self, message: str):
        """Send message to all connected clients"""
        for connection in self.active_connections:
            await connection.send_text(message)
    
    def add_to_history(self, message: Dict):
        """Add message to history"""
        self.message_history.append(message)
        # Keep only last 100 messages
        if len(self.message_history) > 100:
            self.message_history = self.message_history[-100:]

# Create connection manager
manager = ConnectionManager()

# WebSocket LLM
class WebSocketLLM:
    """LLM for WebSocket chat"""
    
    def __init__(self):
        self.conversation_history = []
    
    async def generate_response(self, user_message: str) -> str:
        """Generate response to user message"""
        # Add to conversation history
        self.conversation_history.append({
            "role": "user",
            "content": user_message
        })
        
        # Simple response logic
        user_lower = user_message.lower()
        
        if "hello" in user_lower or "hi" in user_lower:
            response = "Hello! How can I assist you today?"
        elif "bye" in user_lower or "goodbye" in user_lower:
            response = "Goodbye! Have a great day!"
        elif "help" in user_lower:
            response = "I'm here to help! You can ask me anything, and I'll do my best to assist you."
        elif "?" in user_message:
            response = f"That's an interesting question about '{user_message}'. Let me provide you with a thoughtful answer."
        else:
            response = f"I understand you're saying: '{user_message}'. Let me help you with that."
        
        # Add to conversation history
        self.conversation_history.append({
            "role": "assistant",
            "content": response
        })
        
        return response

# Create WebSocket LLM
ws_llm = WebSocketLLM()

@app.websocket("/ws/chat")
async def websocket_chat_endpoint(websocket: WebSocket):
    """
    WebSocket endpoint for real-time chat
    
    Message format:
    {
        "type": "message" | "ping" | "history",
        "content": "message content",
        "timestamp": 1234567890
    }
    """
    await manager.connect(websocket)
    
    try:
        # Send connection confirmation
        await websocket.send_json({
            "type": "system",
            "content": "Connected to chat server",
            "timestamp": time.time()
        })
        
        # Send message history
        if manager.message_history:
            await websocket.send_json({
                "type": "history",
                "messages": manager.message_history,
                "timestamp": time.time()
            })
        
        while True:
            # Receive message from client
            data = await websocket.receive_json()
            message_type = data.get("type", "message")
            
            if message_type == "ping":
                # Respond to ping with pong
                await websocket.send_json({
                    "type": "pong",
                    "timestamp": time.time()
                })
            
            elif message_type == "message":
                user_message = data.get("content", "")
                timestamp = data.get("timestamp", time.time())
                
                # Add user message to history
                user_msg = {
                    "role": "user",
                    "content": user_message,
                    "timestamp": timestamp
                }
                manager.add_to_history(user_msg)
                
                # Echo user message back (for confirmation)
                await websocket.send_json({
                    "type": "message",
                    "role": "user",
                    "content": user_message,
                    "timestamp": timestamp
                })
                
                # Generate response
                response = await ws_llm.generate_response(user_message)
                
                # Send assistant response
                assistant_msg = {
                    "type": "message",
                    "role": "assistant",
                    "content": response,
                    "timestamp": time.time()
                }
                
                manager.add_to_history({
                    "role": "assistant",
                    "content": response,
                    "timestamp": assistant_msg["timestamp"]
                })
                
                await websocket.send_json(assistant_msg)
            
            elif message_type == "history":
                # Send message history
                await websocket.send_json({
                    "type": "history",
                    "messages": manager.message_history,
                    "timestamp": time.time()
                })
    
    except WebSocketDisconnect:
        manager.disconnect(websocket)
    except Exception as e:
        print(f"WebSocket error: {e}")
        manager.disconnect(websocket)

print("=" * 60)
print("WebSocket 后端演示")
print("=" * 60)
print("\nWebSocket 端点已创建！")
print("\n端点：")
print("  WS /ws/chat - WebSocket 聊天")
print("\n消息类型：")
print("  - message: 用户消息")
print("  - ping: 心跳检测")
print("  - history: 获取历史消息")
print("\n✓ WebSocket 后端实现完成！")
````

````python
# 🔬 Micro Practice: SSE Streaming Backend
# Goal: Implement FastAPI endpoint with Server-Sent Events
# Expected outcome: Token-by-token streaming response

from fastapi import FastAPI
from fastapi.responses import StreamingResponse
from fastapi.middleware.cors import CORSMiddleware
import asyncio
from typing import AsyncGenerator

# Create FastAPI app
app = FastAPI()

# Add CORS middleware
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Simulated streaming LLM
class StreamingLLM:
    """LLM that generates tokens one by one"""
    
    def __init__(self):
        self.responses = {
            "hello": "Hello! How can I assist you today?",
            "weather": "I don't have access to real-time weather data, but I can help you with other questions!",
            "default": "That's an interesting question. Let me think about it carefully and provide you with a detailed response."
        }
    
    async def generate_stream(self, prompt: str) -> AsyncGenerator[str, None]:
        """Generate response token by token"""
        # Select response based on prompt
        if "hello" in prompt.lower() or "hi" in prompt.lower():
            response = self.responses["hello"]
        elif "weather" in prompt.lower():
            response = self.responses["weather"]
        else:
            response = self.responses["default"]
        
        # Stream tokens (words in this case)
        words = response.split()
        for i, word in enumerate(words):
            # Add space before word (except first)
            token = word if i == 0 else f" {word}"
            yield token
            # Simulate generation delay
            await asyncio.sleep(0.1)

# Create LLM instance
streaming_llm = StreamingLLM()

@app.post("/api/stream")
async def stream_response(request: dict):
    """
    Stream LLM response using Server-Sent Events
    
    Request body:
    {
        "prompt": "user message"
    }
    """
    prompt = request.get("prompt", "")
    
    async def event_generator():
        """Generate SSE events"""
        try:
            async for token in streaming_llm.generate_stream(prompt):
                # SSE format: data: {content}\n\n
                yield f"data: {json.dumps({'token': token})}\n\n"
            
            # Send completion event
            yield f"data: {json.dumps({'done': True})}\n\n"
        except Exception as e:
            # Send error event
            yield f"data: {json.dumps({'error': str(e)})}\n\n"
    
    return StreamingResponse(
        event_generator(),
        media_type="text/event-stream",
        headers={
            "Cache-Control": "no-cache",
            "Connection": "keep-alive",
        }
    )

@app.get("/api/health")
async def health_check():
    """Health check endpoint"""
    return {"status": "healthy"}

print("=" * 60)
print("SSE 流式后端演示")
print("=" * 60)
print("\nFastAPI 应用已创建！")
print("\n端点：")
print("  POST /api/stream - 流式生成响应")
print("  GET  /api/health - 健康检查")
print("\n✓ SSE 后端实现完成！")
````

## 4. 流式响应 (Streaming Responses)

### 4.1 为什么需要流式响应？

**传统请求-响应的问题**：
- 用户需要等待完整响应（可能 10-30 秒）
- 无法显示生成进度
- 无法提前取消
- 用户体验差

**流式响应的优势**：
- **即时反馈**：立即看到第一个 token
- **降低感知延迟**：从 30 秒降到 < 1 秒
- **可中断**：用户可以提前停止
- **更好的 UX**：类似打字效果

### 4.2 Server-Sent Events (SSE)

**SSE 协议**：
```
HTTP/1.1 200 OK
Content-Type: text/event-stream
Cache-Control: no-cache
Connection: keep-alive

data: Hello
data: world
data: !
```

**SSE 特点**：
- 基于 HTTP
- 单向（服务器→客户端）
- 自动重连
- 文本格式
- 简单易用

**SSE vs WebSocket**：

| 特性 | SSE | WebSocket |
|------|-----|-----------|
| **方向** | 单向 | 双向 |
| **协议** | HTTP | WS |
| **重连** | 自动 | 手动 |
| **复杂度** | 低 | 中 |
| **适用场景** | 文本生成 | 实时聊天 |

### 4.3 SSE 数据格式

**基本格式**：
```
data: message content\n\n
```

**带 ID**：
```
id: 123
data: message content\n\n
```

**多行数据**：
```
data: line 1
data: line 2
data: line 3\n\n
```

**JSON 数据**：
```
data: {"type": "token", "content": "Hello"}\n\n
```

## 2. 前端架构基础 (Frontend Architecture)

### 2.1 LLM 前端通信模式

**1. 请求-响应（Request-Response）**
```
客户端 → [HTTP POST] → 服务器
客户端 ← [完整响应] ← 服务器
```
- **优点**：简单、兼容性好
- **缺点**：阻塞、无法显示进度
- **适用**：短文本生成、批处理

**2. Server-Sent Events (SSE)**
```
客户端 → [HTTP GET] → 服务器
客户端 ← [token 1] ← 服务器
客户端 ← [token 2] ← 服务器
客户端 ← [token 3] ← 服务器
```
- **优点**：单向流式、自动重连、简单
- **缺点**：仅服务器→客户端
- **适用**：文本生成、实时更新

**3. WebSocket**
```
客户端 ↔ [双向通信] ↔ 服务器
```
- **优点**：双向、低延迟、持久连接
- **缺点**：复杂、需要心跳保活
- **适用**：多轮对话、协作编辑

**4. 轮询（Polling）**
```
客户端 → [定期请求] → 服务器
客户端 ← [状态更新] ← 服务器
```
- **优点**：兼容性最好
- **缺点**：延迟高、资源浪费
- **适用**：降级方案

### 2.2 技术选型对比

| 特性 | SSE | WebSocket | 轮询 |
|------|-----|-----------|------|
| **延迟** | 低 | 极低 | 高 |

| **自动重连** | ✅ | ❌ | ✅ |

| **实现复杂度** | 简单 | 中等 | 简单 |

### 问题 1：如何为 LLM 应用构建流畅的用户界面？流式响应（Streaming）对用户体验的提升体现在哪些方面？

**答：**

**流畅用户界面的核心设计原则：**

1. **即时反馈**：用户发送消息后，立即显示"正在生成..."的状态指示
2. **流式显示**：逐字/逐词展示 LLM 的输出，而非等待完整响应后一次性显示
3. **响应式布局**：适配不同屏幕尺寸，保证移动端体验
4. **错误处理**：网络中断、超时等异常情况要有优雅的提示和重试机制

**流式响应的用户体验优势：**

| 对比维度 | 非流式响应 | 流式响应 |
|---------|-----------|---------|
| **首字时间（TTFT）** | 等待完整响应（数秒甚至数十秒）| 几百毫秒内开始显示 |
| **感知延迟** | 高（用户感觉系统"卡住"）| 低（用户看到实时生成）|
| **任务放弃率** | 高（长时间等待导致用户离开）| 低（持续的视觉反馈维持参与感）|
| **交互自然度** | 低（类似批处理）| 高（类似真人打字的体验）|

**从技术角度看流式的实现路径：**

```
LLM 推理引擎（Token by Token 生成）
    ↓
后端流式 API（yield / async generator）
    ↓
HTTP 流式传输（SSE 或 WebSocket）
    ↓
前端逐步追加显示（innerHTML append / React state update）
```

**关键优化点**：前端接收到 token 后应批量更新 DOM（每 50ms 合并一次更新），避免频繁重绘造成的性能问题。

### 问题 2：Server-Sent Events（SSE）和 WebSocket 各有什么特点？在 LLM 聊天应用中应如何选择？

**答：**

**技术特性对比：**

| 特性 | SSE | WebSocket |
|-----|-----|-----------|
| **通信方向** | 单向（服务器 → 客户端）| 双向（全双工）|
| **协议** | 基于 HTTP/1.1 | 独立的 WS 协议（HTTP 握手升级）|
| **自动重连** | 浏览器内置支持 | 需要手动实现 |
| **代理/防火墙兼容性** | 好（标准 HTTP）| 较差（部分防火墙会拦截）|
| **数据格式** | 文本（UTF-8）| 文本或二进制 |
| **服务器资源** | 较少（HTTP 连接）| 较多（持久连接）|
| **实现复杂度** | 低 | 中等 |

**LLM 应用场景选择指南：**

```
场景判断树：
是否需要客户端向服务器发送实时数据？
├── 否（仅流式接收 LLM 输出）→ 选择 SSE
│   适合：简单聊天、文章生成、代码补全
└── 是（需要双向实时通信）→ 选择 WebSocket
    适合：实时协作编辑、多用户聊天室、语音转文字
```

**实践建议：**

- **快速原型 / 内部工具**：优先选 SSE，实现简单，5-10 行代码即可完成
- **生产级聊天应用**：考虑 WebSocket，支持打断（用户停止生成）、多模态（发送图片）等高级功能
- **边缘场景**：若需要在 HTTP/2 多路复用场景下降低连接数，SSE 有天然优势

**SSE 的最大优势**：浏览器自动处理重连（`EventSource` API），服务端实现只需标准 HTTP 即可，与 CDN、负载均衡器的兼容性更好。

### 问题 3：Streamlit 和 Gradio 分别适合哪些场景？与 React 生产级应用相比，它们的局限性在哪里？

**答：**

**工具定位对比：**

| 工具 | 目标用户 | 开发速度 | 自定义能力 | 生产就绪 |
|-----|---------|---------|----------|---------|
| **Gradio** | ML 研究者、模型演示 | 极快（几行代码）| 低 | 有限（内置分享链接）|
| **Streamlit** | 数据科学家、内部工具 | 快（纯 Python）| 中 | 中等 |
| **React + FastAPI** | 前端工程师、生产应用 | 慢（需要全栈知识）| 极高 | 完整 |

**Streamlit 适用场景：**
- 数据探索和可视化仪器盘
- 内部工具（数据标注、模型评估界面）
- 快速验证产品概念（PoC）
- Python 团队无前端资源时

**Gradio 适用场景：**
- 模型 Demo 页面（Hugging Face Spaces）
- 学术论文配套演示
- API 接口可视化测试
- 快速生成可分享的链接

**与 React 生产应用的核心差距：**

```
局限性分析：
1. 性能：Streamlit 每次用户交互触发整个脚本重新执行（全量刷新）
```

**技术选型决策树：**
- 需要快速演示 → Gradio
- 内部数据工具，纯 Python 团队 → Streamlit
- 面向用户的生产应用，需要高并发、自定义 UI → React + FastAPI

### 问题 4：如何在前端正确渲染 LLM 生成的富文本内容（Markdown、代码高亮、LaTeX 数学公式）？需要注意哪些安全风险？

**答：**

**富文本渲染技术栈：**

| 内容类型 | 推荐库 | 示例 |
|---------|-------|------|
| **Markdown** | `marked.js` / `react-markdown` | `**粗体**` → `<strong>粗体</strong>` |
| **代码高亮** | `highlight.js` / `Prism.js` | 自动识别语言并着色 |
| **LaTeX 数学** | `KaTeX` / `MathJax` | `$E=mc^2$` → 渲染为数学公式 |
| **表格** | Markdown 扩展 + CSS | GFM（GitHub Flavored Markdown）|

**渲染流水线（React 示例）：**

```
LLM 输出（原始文本）
    ↓ marked.js（Markdown → HTML）
    ↓ DOMPurify（XSS 净化）← 安全关键步骤！
    ↓ KaTeX（LaTeX 渲染）
    ↓ highlight.js（代码着色）
    ↓ dangerouslySetInnerHTML（React 注入）
```

**安全风险与防护措施：**

**XSS（跨站脚本攻击）**是最主要风险：

```javascript
// 危险：直接注入未经净化的 HTML
element.innerHTML = markedOutput;  // XSS 漏洞！

// 安全：使用 DOMPurify 净化
import DOMPurify from 'dompurify';
element.innerHTML = DOMPurify.sanitize(markedOutput, {
  ALLOWED_TAGS: ['p', 'br', 'strong', 'em', 'code', 'pre', 'table'],
  ALLOWED_ATTR: ['class', 'href']
});
```

**关键防护原则：**

1. **永远净化 HTML**：使用 DOMPurify 或 sanitize-html，绝不相信 LLM 输出是安全的
2. **限制标签白名单**：仅允许展示性 HTML 标签，禁止 `<script>`、`<iframe>`、`onclick` 等
3. **CSP（内容安全策略）**：通过 HTTP 头部限制可执行脚本来源
4. **流式渲染安全**：逐步接收时，需等 Markdown 块完整再渲染（避免中间状态的不完整标签注入）

### 问题 5：LLM 前端应用的全栈部署架构是什么样的？如何选择合适的部署方案？从开发到生产需要考虑哪些关键因素？

**答：**

**生产环境部署架构：**

```
用户设备
    ↓ HTTPS
CDN（Cloudflare / AWS CloudFront）
    ├── 静态资源缓存（HTML/CSS/JS）
    └── 动态请求转发
            ↓
    负载均衡器（Nginx / AWS ALB）
            ↓
    前端服务（Vercel / Netlify）      后端集群（多实例）
    ├── React 静态文件         ←→     ├── FastAPI 推理服务
    └── 构建优化（Code Split）         ├── 数据库（PostgreSQL）
                                      ├── 缓存（Redis）
                                      └── 对象存储（S3）
```

**部署方案选型：**

| 场景 | 前端 | 后端 | 成本 | 复杂度 |
|-----|------|------|------|-------|
| **个人项目** | Vercel / GitHub Pages | Railway / Render | 免费-低 | 低 |
| **初创公司** | Vercel + CDN | Fly.io + 托管数据库 | 中 | 中 |
| **企业级** | AWS S3 + CloudFront | AWS ECS/EKS | 高 | 高 |
| **私有部署** | Nginx 自托管 | Docker Compose | 服务器成本 | 中高 |

**从开发到生产的关键考量清单：**

1. **性能优化**：
   - 代码分割（Code Splitting）减少首屏加载时间
   - LLM API 响应缓存（相同查询不重复请求）
   - 静态资源 CDN 加速

2. **可靠性**：
   - 流式响应断线重连机制
   - API 调用超时设置（建议 30-60 秒）
   - 熔断器模式（Circuit Breaker）防止级联故障

3. **可观测性**：
   - 前端错误监控（Sentry）
   - API 调用链路追踪（OpenTelemetry）
   - 用户行为分析（请求延迟、会话长度）

4. **安全性**：
   - API Key 绝不暴露在前端代码中（通过后端中转）
   - 用户认证（JWT / OAuth 2.0）
   - 速率限制（Rate Limiting）防止滥用

5. **成本控制**：
   - LLM Token 用量监控和告警
   - 用户级别的配额管理
   - 对话历史压缩，减少每次请求的 token 数
